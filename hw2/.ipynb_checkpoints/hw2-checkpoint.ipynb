```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import warnings
from sklearn.linear_model import SGDClassifier
```

```python
df = pd.read_csv('hw2.csv', sep=',', header=0, usecols='V')
cf = pd.read_csv('hw2.csv', sep=',', header=0, usecols='C')
```

```python
tr_set = np.array(df[:40])
test_set = np.array(df[40:])
tr_target = np.array(cf[:40])
test_target = np.array(cf[40:])
data_set = np.array(df)
data_class = np.array(cf)
```

```python
plt.scatter(data_set, data_class , marker="x", color="blue")
plt.title("The plot of the given example dataset")
plt.ylim(-1, +2)
plt.xlim(1, 7)
plt.xlabel("x")
plt.ylabel("Class Code")
```

```python
epoch_list = [5, 10, 25, 50, 100, 500]
for ep_value in epoch_list:
    sgd = SGDClassifier(n_iter=ep_value, loss="log", learning_rate="optimal")
    with warnings.catch_warnings():
        warnings.simplefilter("ignore")
        sgd_fit = sgd.partial_fit(tr_set, tr_target, np.unique(tr_target))
        sgd_predict = sgd_fit.predict(test_set)
        
        plt.scatter(tr_set,  sgd_predict, marker="x", color="blue")
        plt.title("The plot of the given example training set")
        plt.ylim(-1, +2)
        plt.xlim(1, 7)
        plt.xlabel("x")
        plt.ylabel("P(C | X)")
```

```python
epoch_list = [5, 10, 25, 50, 100, 500]
for ep_value in epoch_list:
    sgd = SGDClassifier(n_iter=ep_value, loss="log", learning_rate="optimal")
    with warnings.catch_warnings():
        warnings.simplefilter("ignore")
        sgd_fit = sgd.partial_fit(tr_set, tr_target, np.unique(tr_target))
        sgd_predict = sgd_fit.predict(test_set)
        
        plt.scatter(tr_set,  sgd_predict, marker="x", color="blue")
        plt.scatter(test_set,  sgd_predict, marker="o", color="green")
        plt.title("The plot of the given example training and test set")
        plt.ylim(-1, +2)
        plt.xlim(1, 7)
        plt.xlabel("x")
        plt.ylabel("P(C | X)")
```
